### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1= pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Primero sacamos cual tienda es la que mas vende y cual es la que menos vende!

```python
ingreso1 = tienda1['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()
```

```python
print("Ingreso total Tienda 1:", ingreso1)
print("Ingreso total Tienda 2:", ingreso2)
print("Ingreso total Tienda 3:", ingreso3)
print("Ingreso total Tienda 4:", ingreso4)
```

```python
import matplotlib.pyplot as plt
```

```python
ingresos = [ingreso1, ingreso2, ingreso3, ingreso4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Aca podemos observar que la tienda que mas vende es la tienda 1 y la que menos vende es la 4


# 2. Ventas por categoría

```python
categorias1 = tienda1['Categoría del Producto'].value_counts()
categorias2 = tienda2['Categoría del Producto'].value_counts()
categorias3 = tienda3['Categoría del Producto'].value_counts()
categorias4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print("Categorías más populares en Tienda 1:\n", categorias1)
print("\nCategorías más populares en Tienda 2:\n", categorias2)
print("\nCategorías más populares en Tienda 3:\n", categorias3)
print("\nCategorías más populares en Tienda 4:\n", categorias4)
```

```python
import matplotlib.pyplot as plt

# Crear listas con los datos
categorias = categorias1.index.tolist()
ventas1 = categorias1.values
ventas2 = categorias2.reindex(categorias, fill_value=0).values
ventas3 = categorias3.reindex(categorias, fill_value=0).values
ventas4 = categorias4.reindex(categorias, fill_value=0).values

# Crear la figura
plt.figure(figsize=(10,6))

# Graficar cada tienda
plt.plot(categorias, ventas1, label='Tienda 1', marker='o')
plt.plot(categorias, ventas2, label='Tienda 2', marker='s')
plt.plot(categorias, ventas3, label='Tienda 3', marker='^')
plt.plot(categorias, ventas4, label='Tienda 4', marker='d')

# Personalizar gráfico
plt.title('Cantidad de productos vendidos por categoría')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio1 = tienda1['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()
```

```python
print("Costo de envío promedio - Tienda 1:", envio1)
print("Costo de envío promedio - Tienda 2:", envio2)
print("Costo de envío promedio - Tienda 3:", envio3)
print("Costo de envío promedio - Tienda 4:", envio4)
```

```python
import matplotlib.pyplot as plt

costos = [envio1, envio2, envio3, envio4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8,5))
plt.bar(tiendas, costos, color='salmon')
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de envío promedio')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
